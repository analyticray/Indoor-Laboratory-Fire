```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#import pandas_datareader as pdr
import numpy as np
import datetime

from pandas import read_csv
from pandas import datetime
```

```python
series = read_csv('electrical_1.csv', header=0, index_col=0, squeeze=True)
series = series.drop('Reading ID', axis=1)
series = series.drop('Status', axis=1)
print(series.head())
```

```python
myBasicCorr = series.corr()
sns.heatmap(myBasicCorr)
```

```python
sns.heatmap(myBasicCorr, annot = True)
```

```python
akws = {"ha": 'left',"va": 'top'}
```

```python
ax = sns.heatmap(myBasicCorr, annot=True, annot_kws=akws)
```

```python
for p in ax.texts:
    myTrans = p.get_transform()
    offset = matplotlib.transforms.ScaledTranslation(-0.45, 0.45, matplotlib.transforms.IdentityTransform())
    p.set_transform(myTrans + offset)
```

```python
mask = np.zeros(myBasicCorr.shape, dtype=bool)
mask[np.triu_indices(len(mask))] = True
sns.heatmap(myBasicCorr, vmin = -1, vmax = 1, center = 0, cmap = 'coolwarm', annot = True, mask = mask)
```
